```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')

print('Matplotlib version: ', mpl.__version__)
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
```

```python
df_can.head()
```

```python
print(df_can.shape)
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.head()
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
years
```

```python
df_continents = df_can.groupby('Continent', axis=0).sum()

print(type(df_can.groupby('Continent', axis=0)))
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(5, 6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True)

plt.title('Immigration to Canada by Continent [1980-2013]')
plt.axis('equal') # makes pie chart look like a circle
plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge

df_continents['Total'].plot(kind='pie',
                           figsize=(10, 6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True,
                           labels=None,
                           pctdistance=1.12,
                           colors=colors_list,
                           explode=explode_list)

plt.title('Immigration to Canada by Continent 1980 - 2013', y=1.12, fontsize=15)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left', fontsize=8)
plt.show()
```

```python
explode_list = [0.0, 0, 0, 0.1, 0.1, 0.2]

df_continents['2013'].plot(kind='pie',
                          figsize=(15,6),
                          autopct='%1.1f%%',
                          startangle=90,
                          shadow=True,
                          labels=None,
                          pctdistance=1.12,
                          explode=explode_list)

# scal the title up by 12% to match pctdistance
plt.title('Immigration to Canada by Continent in 2013', y=1.12)
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')
plt.show()
```

# A box plot to visualize the distribution of the top 15 countries grouped by decades (1980s, 1990s, 2000s)

```python
df_top15 = df_can.sort_values(['Total'], ascending=False, axis=0).head(15)
df_top15
```

# New DataFrame which contains the aggregate by decade

```python
##### List of all years in each decade
years_80s = list(map(str, range(1980, 1990)))
years_90s = list(map(str, range(1990, 2000)))
years_00s = list(map(str, range(2000, 2010)))

# original dataframe slice / creat a series for each decade
df_80s = df_top15.loc[:, years_80s].sum(axis=1)
df_90s = df_top15.loc[:, years_90s].sum(axis=1)
df_00s = df_top15.loc[:, years_00s].sum(axis=1)

# merge all the series into a df
aggregate_df = pd.DataFrame({'1980s': df_80s, '1990s': df_90s, '2000s': df_00s}) 

aggregate_df.head()
```

```python
aggregate_df.describe()
```

# Box plot of new df

```python
aggregate_df.plot(kind='box', figsize=(10,6))
plt.title('Immigration of top 15 countires grouped by decade')
plt.show()
```

# Scatter plot stuff

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

# change years to int
df_tot.index = map(int, df_tot.index)

# reset the index to put back in as a column in the df_to df
df_tot.reset_index(inplace=True)

# rename columns
df_tot.columns = ['year', 'total']

# view the final dataframe
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10,6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
x = df_tot['year']
y = df_tot['total']
fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Num of immigrants')

# line best fit plot
plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 15000))

plt.show()
```

# Scatter plot of the total immigration from Denmark, Norway, and Sweden to Canada from 1980 to 2013

```python
df_countries = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()

# create df that sums 3 countries for each year
df_total = pd.DataFrame(df_countries.sum(axis=1))

# reset index in place
df_total.reset_index(inplace=True)

# rename columns
df_total.columns = ['year', 'total']

# change column year from string to int to create scatter plot
df_total['year'] = df_total['year'].astype(int)

df_total.head()
```

```python
df_total.plot(kind='scatter', x='year', y='total', figsize=(10,6), color='darkblue')

plt.title('Immigration from Denmark, Norway, and Sweden to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
df_arg = df_can[years].transpose()

# cast the years (the index) to type int
df_arg.index = map(int, df_arg.index)

# label index
df_arg.index.name = 'Year'

# reset index to bring the Year in as a column
df_arg.reset_index(inplace=True)

df_arg.head()
```

```python
# normalize Brazil data
normal_brazil = (df_arg['Brazil'] - df_arg['Brazil'].min()) / (df_arg['Brazil'].max() - df_arg['Brazil'].min())

# normalize Argentina data
normal_argentina = (df_arg['Argentina'] - df_arg['Argentina'].min()) / (df_arg['Argentina'].max() - df_arg['Argentina'].min())
```

# Bubble Plot

```python
ax0 = df_arg.plot(kind='scatter',
                 x='Year',
                 y='Brazil',
                 figsize=(14, 8),
                 alpha=0.5,
                 color='green',
                 s=normal_brazil * 2000 + 10,  # pass in weights
                 xlim=(1975, 2015))

ax1 = df_arg.plot(kind='scatter',
                  x='Year',
                  y='Argentina',
                  alpha=0.5,
                  color='blue',
                  s=normal_argentina * 2000 + 10,
                  ax=ax0)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 to 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

# Do the same for China , India

```python
norm_china = (df_arg['China'] - df_arg['China'].min()) / (df_arg['China'].max() - df_arg['China'].min())

norm_india = (df_arg['India'] - df_arg['India'].min()) / (df_arg['India'].max() - df_arg['India'].min())
```

```python
ax0 = df_arg.plot(kind='scatter',
                 x='Year',
                 y='China',
                 figsize=(14, 8),
                 alpha=0.5,
                 color='green',
                 s=norm_china * 2000 + 10,  # pass in weights
                 xlim=(1975, 2015))

ax1 = df_arg.plot(kind='scatter',
                  x='Year',
                  y='India',
                  alpha=0.5,
                  color='blue',
                  s=norm_india * 2000 + 10,
                  ax=ax0)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from China and India from 1980 to 2013')
ax0.legend(['China', 'India'], loc='upper left', fontsize='x-large')
```
